```python
import matplotlib.pyplot as plt
```

```python
#Histogram of ages, separating survived and not survived

# Separate survivors and non-survivors
survived = df[df['Survived'] == 1]
not_survived = df[df['Survived'] == 0]

# Calculate mean and median ages before replacing missing values
mean_age_survived = survived['Age'].mean()
median_age_survived = survived['Age'].median()
mean_age_not_survived = not_survived['Age'].mean()
median_age_not_survived = not_survived['Age'].median()

print("Mean and Median Ages (before replacing missing values):")
print(f"Survived - Mean: {mean_age_survived:.2f}, Median: {median_age_survived:.2f}")
print(f"Not Survived - Mean: {mean_age_not_survived:.2f}, Median: {median_age_not_survived:.2f}")

# Plot age distribution for survivors and non-survivors
plt.figure(figsize=(10, 6))

# Plot for survivors
sns.histplot(survived['Age'], kde=True, bins=30, color='green', label='Survived', alpha=0.6)

# Plot for non-survivors
sns.histplot(not_survived['Age'], kde=True, bins=30, color='red', label='Not Survived', alpha=0.3)

# Add titles and labels
plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Replace missing ages with mean age

# Calculate the mean age, excluding missing values
mean_age = df['Age'].mean()

# Replace missing values in the Age column with the mean age
df['Age'].fillna(mean_age, inplace=True)

# Verify if there are any remaining missing values
print(df['Age'].isnull().sum())  # Should print 0 if all missing values are replaced
```

```python
#Histogram of ages, separating survived and not survived (with replaced ages - skews graph)

# Separate survivors and non-survivors
survived = df[df['Survived'] == 1]
not_survived = df[df['Survived'] == 0]

# Calculate mean and median ages before replacing missing values
mean_age_survived = survived['Age'].mean()
median_age_survived = survived['Age'].median()
mean_age_not_survived = not_survived['Age'].mean()
median_age_not_survived = not_survived['Age'].median()

print("Mean and Median Ages (before replacing missing values):")
print(f"Survived - Mean: {mean_age_survived:.2f}, Median: {median_age_survived:.2f}")
print(f"Not Survived - Mean: {mean_age_not_survived:.2f}, Median: {median_age_not_survived:.2f}")

# Plot age distribution for survivors and non-survivors
plt.figure(figsize=(10, 6))

# Plot for survivors
sns.histplot(survived['Age'], kde=True, bins=30, color='green', label='Survived', alpha=0.6)

# Plot for non-survivors
sns.histplot(not_survived['Age'], kde=True, bins=30, color='red', label='Not Survived', alpha=0.3)

# Add titles and labels
plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
